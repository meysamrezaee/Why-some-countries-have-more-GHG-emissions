```python
correctionList = [
('Bolivia (Plurin. State of)', 'Bolivia'),
('CÃ´te dâ\x80\x99Ivoire', 'Côte d\'Ivoire'),
('Côte d\x92Ivoire', 'Côte d\'Ivoire'),
('Cote d\'Ivoire', 'Côte d\'Ivoire'),
('Czechia', 'Czechia (Czech Republic)'),
('Czech Republic', 'Czechia (Czech Republic)'),
('Dem. Rep. of the Congo', 'Democratic Republic of the Congo'),
('Democratic Republic of Congo', 'Democratic Republic of the Congo'),
('Eswatini', 'Eswatini (fmr. "Swaziland")'),
('Swaziland', 'Eswatini (fmr. "Swaziland")'),
('Iran (Islamic Republic of)', 'Iran'),
('Russian Federation', 'Russia'),
('Saint Vincent & Grenadines', 'Saint Vincent and the Grenadines'),
('United Rep. of Tanzania', 'Tanzania'),
('Venezuela (Boliv. Rep. of)', 'Venezuela'),
('Congo', 'Congo (Congo-Brazzaville)'),
('Viet Nam', 'Vietnam'),
('State of Palestine', 'Palestine State'),
('Sudan [former]', 'Sudan'),
('United States', 'United States of America'),
('Palestine', 'Palestine State')
]
```

```python
def transformName(name):
    for i in correctionList:
        if name == i[0]:
            return i[1]
    return 'NA'
```

```python
# go through a dataframe and check every item for transformation
def fixCountryName(initDataFrame):
    for j in range(initDataFrame.shape[1]):
        if initDataFrame.iloc[:, j].dtype == np.int64 or initDataFrame.iloc[:, j].dtype == np.float64:
            continue
        for i in range(initDataFrame.shape[0]):
            if str(initDataFrame.iloc[i, j]).lower() == 'nan':
                continue
            if type(initDataFrame.iloc[i, j]) != np.int64 and type(initDataFrame.iloc[i, j]) != np.float64:
                newValue = transformName(initDataFrame.iloc[i, j])
            if newValue != 'NA':
                initDataFrame.iloc[i, j] = newValue
```

```python
# Fix the country names in all dataframes
dfList = [populationDF, urbanDF, gdpDF, exchangeRateDF, FertilityDF, productionDF, 
 oilConsumptionDF, gasConsumptionDF, coalConsumptionDF, ghgEmissionDF]
for i in dfList:
    fixCountryName(i)
```

### Initialize the data frame by the country column

```python
df = pd.DataFrame(actualNames)
df.columns = ['Country']
```

```python
# newDataFrame must have two columns, first the country name, second the values of the new variable
# newColumnName is the name of the new column in df
```

```python
def addToDataFrame(newDataFrame, newColumnName):
    df[newColumnName] = [np.NaN for i in range(df.shape[0])]
    newColumnIndex = df.columns.get_loc(newColumnName)
    countryColumnName = newDataFrame.columns[0]
    for i in range(df.shape[0]):
        name = df.iloc[i, 0]
        newData = newDataFrame[newDataFrame[countryColumnName] == name]
        if newData.shape[0] > 0:
            df.iloc[i, newColumnIndex] = newData.iloc[0, 1]
```
